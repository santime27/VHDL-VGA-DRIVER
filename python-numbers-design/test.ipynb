```python
from PIL import Image
import matplotlib.pyplot as plt

def split_image_to_list(image_path, cell_width, cell_height, rows):
    """
    Divide una imagen en celdas y las almacena en una lista de variables.
    
    Parámetros:
    - image_path: Ruta a la imagen PNG.
    - cell_width: Ancho de cada celda/caracter.
    - cell_height: Alto de cada celda/caracter.
    - columns: Número de columnas en la imagen.
    - rows: Número de filas en la imagen.
    
    Retorna:
    - Una lista de imágenes recortadas (caracteres).
    """
    
    # Cargar la imagen
    image = Image.open(image_path)
    image_width, image_height = image.size

    # Lista para almacenar los caracteres
    char_list = []

    # Cortar la imagen en celdas y almacenar en la lista
    for alto in range(0,image_height,cell_height):
        for col in range(0, image_width, cell_width):
            # Calcular las coordenadas de la celda actual
            #left, upper, right, lower
            left = col
            upper = alto
            right = col + cell_width
            lower = alto + cell_height

            # Recortar la imagen en la posición especificada
            cropped_image = image.crop((left, upper, right, lower))
            plt.figure(figsize=(3, 5))
            plt.imshow(cropped_image)
            plt.axis('off')
            plt.show()
            plt.show()

            
            # Añadir la imagen recortada a la lista
            char_list.append(cropped_image)

    return char_list

# Parámetros de entrada
image_path      = "fuente.png"
#alto y ancho de la imagen en pixeles
cell_width      = 70   # Ancho de cada letra (ajústalo según tu imagen)
cell_height     = 100  # Alto de cada letra (ajústalo según tu imagen)
rows            = 4    # Número de filas en tu imagen

# Llamar a la función para obtener la lista de caracteres
char_images = split_image_to_list(image_path, cell_width, cell_height, rows)

# Verificar el contenido del vector
print(f"Total characters extracted: {len(char_images)}")

# Ejemplo: Mostrar el primer carácter extraído (opcional)
plt.figure(figsize=(3, 5))
plt.imshow(char_images[3])
plt.axis('off')
plt.show()
plt.show()
```
